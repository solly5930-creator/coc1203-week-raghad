```python
import math

# Function to calculate area
def calculate_area(radius):
    area = math.pi * (radius ** 2)
    return round(area, 2)

# Lambda to calculate circumference
calculate_circumference = lambda radius: round(2 * math.pi * radius, 2)

# User input
radius = float(input("Enter the radius of the circle: "))

# Calculations
area = calculate_area(radius)
circumference = calculate_circumference(radius)

# Output
print("Area:", area)
print("Circumference:", circumference)
```

```python
import re

# Function to sum numbers
def sum_numbers(numbers):
    total = 0
    for num in numbers:
        total += int(num)
    return total

# Lambda to double the sum
double_value = lambda x: x * 2

# User input
text = input("Enter a string: ")

# Find numbers using regex
numbers = re.findall(r'\d+', text)

# Calculations
total_sum = sum_numbers(numbers)
doubled = double_value(total_sum)

# Output
print("Numbers found:", numbers)
print("Sum:", total_sum)
print("Doubled value:", doubled)
```

```python
import re
import math

# Function to count vowels
def count_vowels(text):
    vowels = re.findall(r'[aeiouAEIOU]', text)
    return len(vowels)

# Lambda for word length
word_length = lambda word: len(word)

# User input
paragraph = input("Enter a paragraph: ")

# Find words starting with vowel
vowel_words = re.findall(r'\b[aeiouAEIOU]\w*', paragraph)

# Count vowels
total_vowels = count_vowels(paragraph)

# Calculate average word length
if vowel_words:
    lengths = [word_length(word) for word in vowel_words]
    average_length = round(sum(lengths) / len(lengths), 2)
else:
    average_length = 0

# Output
print("Words starting with vowels:", vowel_words)
print("Total vowels:", total_vowels)
print("Average vowel-word length:", average_length)
```
